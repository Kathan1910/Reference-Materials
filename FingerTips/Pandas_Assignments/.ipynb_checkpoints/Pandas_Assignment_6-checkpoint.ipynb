### 12) Join the two tables user_usage and user_device on use_id column using right join.

### 15) Join the two tables user_usage and user_device on use_id column using outer join.

```python
merge_3 = pd.merge(user_usage,user_device[['use_id', 'platform', 'device']], 
                   on= 'use_id',how='outer')
```
