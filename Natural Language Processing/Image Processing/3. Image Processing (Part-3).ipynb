```python
import cv2
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# Load the predicted satellite image
image = cv2.imread('/home/kathan/Kathan/Computer Vision/Indian AI Production/Images/buildings.png')

# Load the binary mask of the predicted image
binary_mask = cv2.imread('/home/kathan/Kathan/Computer Vision/Indian AI Production/Images/buildings_mask.png', 0)

# Find the contours of houses in the binary mask
contours, hierarchy = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Convert each contour to a polygon
polygons = []
for contour in contours:
    polygon = Polygon(contour.reshape(-1, 2))
    polygons.append(polygon)

# Plot the image and the polygons
fig, ax = plt.subplots()
ax.imshow(image)

for polygon in polygons:
    patch = plt.Polygon(polygon.exterior.xy, edgecolor='green', linewidth=2, fill=False)
    ax.add_patch(patch)

plt.show()
```
